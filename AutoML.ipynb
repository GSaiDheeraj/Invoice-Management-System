```python
import pandas as pd
```

```python
df = pd.read_csv("H2HBABBA2945.csv")
```

```python
df.drop('area_business',axis='columns', inplace=True) # droping area_business coloumn because of having null values 
df.drop('document_create_date.1',axis='columns', inplace=True) # droping area_business coloumn because of having null values 
df.drop('posting_id',axis='columns', inplace=True) # droping area_business coloumn because of having null values 
df.drop('document type',axis='columns', inplace=True) # droping area_business coloumn because of having null values 
```

```python
df = df.dropna(how='any', axis=0) # Droping rows with any number of null values
```

```python
df.head()
```

```python
import datetime as dt
```

```python
#dividing the year, month,date from the number and changing the format 20000704
def date_format(date):
    x = str(date)
    year = int(x[0:4])
    month = int(x[4:6])
    day = int(x[6:])
    return dt.datetime(year, month, day)
```

```python
df['due_in_date'] = pd.to_numeric(df['due_in_date'], downcast='integer') #changing the float type to integer
```

```python
df['baseline_create_date'] = pd.to_numeric(df['baseline_create_date'], downcast='integer') #changing float to integer
```

```python
df['due_in_date'] = df['due_in_date'].apply(date_format) #changing the integer to year-month-date format
```

```python
df['document_create_date'] = df['document_create_date'].apply(date_format) #changing the integer to year-month-date form
```

```python
df['baseline_create_date'] = df['baseline_create_date'].apply(date_format) #changing integer type to year-month-date form
```

```python
df['clear_date']= pd.to_datetime(df['clear_date']) #changing object type to datetime format
```

```python
df['delay'] = (df['clear_date'] - df['due_in_date']).dt.days
```

```python
df.head()
```

```python
x = df.drop("delay", axis=1)
y = df['delay']
```

```python
import evalml
```

```python
evalml.problem_types.ProblemTypes.all_problem_types
```

```python
X_train,X_test,y_train,y_test=evalml.preprocessing.split_data(x,y,problem_type='regression')
```

```python
from evalml import AutoMLSearch
```

```python
automl=AutoMLSearch(X_train=X_train,y_train=y_train,problem_type='regression',max_batches=1,optimize_thresholds=True)
```

```python
automl.search()
```

```python
automl.rankings
```

```python
automl.best_pipeline
```

```python
best_pipeline = automl.best_pipeline
```

```python
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python
scores = best_pipeline.score(X_test, y_test,  objectives=evalml.objectives.get_core_objectives('regression'))
```
